```python
#In my sprint, I used pandas dataframe as a data source 
# i first created the dataframe from a pandas series that consisted of a dictionary
# I then decided to create a graphical user interface to implement in my project
# Researched different ways and found the best solution to my understanding and that would be to import the pyqt module.
# intergrated the code of data to the skeleton code that was shared from the internet.
# Pandas is a Python library commonly used for data manipulation and analysis. 
# It provides a nice API for loading 2D tabular data from various data sources and performing data analysis on it.


import sys # allows you access functions and anything else defined in the module
from PyQt5 import QtCore, QtGui, QtWidgets #mport the pyQt library that allows me to use the QT frames to buld different type of interactive GUI 
from PyQt5.QtCore import Qt #from pyqt5 lib that imports the class Qt
import pandas as pd #imports the pandas module


class TableModel(QtCore.QAbstractTableModel): # a class is created using the keyword class and is pass a model in the parameter of the class

    def __init__(self, data): #most important method in the class and is passed with 2 attributes 
        super(TableModel, self).__init__()
        self._data = data

    def data(self, index, role):
        if role == Qt.DisplayRole:
            value = self._data.iloc[index.row(), index.column()] # iloc Purely integer-location based indexing for selection by position.
            return str(value)

    def rowCount(self, index): # Returns the number of rows under the given parent. 
        #When the parent is valid it means that rowCount is returning the number of children of parent.
        return self._data.shape[0] # Returns the data to be displayed starting from number 0 index



    def columnCount(self, index): # Returns the number of columns for the children of the given parent.
        return self._data.shape[1] # returns data to be displaced from the number 1 index
    
    def headerData(self, section, orientation, role): #Returns the data for the given role and section in the header with the specified orientation.

        # section is the index of the column/row.
        #For horizontal headers, the section number corresponds to the column number. 
        # Similarly, for vertical headers, the section number corresponds to the row number.
        if role == Qt.DisplayRole:
            if orientation == Qt.Horizontal:
                return str(self._data.columns[section])

            if orientation == Qt.Vertical:
                return str(self._data.index[section])


class MainWindow(QtWidgets.QMainWindow):

    def __init__(self):
        super().__init__()

        self.table = QtWidgets.QTableView()
        
    
        # a dictionary to create a series
        df = pd.DataFrame(columns=["Day-Part", "Start Time", "End Time"])
        parts = int(input("Enter the number of day parts:"))

        for _ in range(parts):
            dp = input("Enter Part of the Day ")
            st = input("Enter start time for {}".format(dp))
            et = input("Enter end time for {}".format(dp))
            df1 = pd.DataFrame(data=[[dp,st,et]],columns=["Day-Part", "Start Time", "End Time"])
            df = pd.concat([df,df1], axis=0)

        df.index = range(len(df.index))
        
        
        self.model = TableModel(df)

        self.table.setModel(self.model)

        self.setCentralWidget(self.table)
        
        title = "Adam Banderker's Dataframe"
  
        # set the title 
        self.setWindowTitle(title) 
        
        self.setGeometry(300, 200 ,1000, 200) 
        self.show()
        
  
app=QtWidgets.QApplication(sys.argv)
window=MainWindow()
window.show()
app.exec_()
```
